```python
df = pd.DataFrame(columns = ['Link','Area','Floor','Rooms','Bathroom','Roommates','Price'])

for link in links_items:
    response = requests.get(link,headers=headers,cookies=cookies)
    #check if the response is without errors
    response.raise_for_status()
    soup = bs4.BeautifulSoup(response.text,'html.parser')
    
    # here i take the features of the house
    section_items = soup.find_all('div',class_ = 'details-property-feature-one')
    features_home = section_items[1].find_all('li')
    
    # here i take the price
    span_price = soup.find('span',class_ = 'info-data-price')
    price = span_price.find('span',class_ = 'txt-bold').text.strip()
    
    length = len(df)
    diz = {}
    for feature in features_home:
        feature = feature.text.strip()
        if 'bagn' in feature:
            diz['Bathroom'] = feature
        elif 'local' in feature:
            diz['Rooms'] = feature
        elif 'Stanz' in feature:
            diz['Area'] = feature
        elif 'Pian' in feature or 'pian' in feature :
            diz['Floor'] = feature
        elif 'persone' in feature:
            diz['Roommates'] = feature
            
    diz['Price'] = price
    diz['Link'] = link
    df.loc[length] = diz
```
